```python
# necessary libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read in .csv file:

```python
# number of responses for each category
responses = data.count()

# bar chart for the response counts
plt.figure(figsize=(12, 6))
responses.plot(kind='bar')
plt.title('Number of Reviews/Responses for Each Category')
plt.ylabel('Number of Responses')
plt.xticks(rotation=45)
plt.show()
```

# Data Cleaning

```python
# option to drop unneccesary columns - already completed
# drop the 'Unnamed: 0' column - index use only
#data_cleaned = data.drop(columns=['Unnamed: 0'])

data_cleaned = data

# convert 'Date Published' and 'Date Flown' to datetime objects
data_cleaned['Date Published'] = pd.to_datetime(data_cleaned['Date Published'])
data_cleaned['Date Flown'] = pd.to_datetime(data_cleaned['Date Flown'], errors='coerce')  # Coerce errors due to invalid parsing

data_cleaned.head()
```

# Impute Values

We built in the option to impute missing values, but decided it ultimately was not necessary for our model.

```python
# option to impute missing values
# fills numerical columns with median values
# fills categorical columns with the mode

'''

numerical_columns = data_cleaned.select_dtypes(include=['float64', 'int64']).columns
categorical_columns = data_cleaned.select_dtypes(include=['object']).columns

for col in numerical_columns:
    data_cleaned[col].fillna(data_cleaned[col].median(), inplace=True)

for col in categorical_columns:
    data_cleaned[col].fillna(data_cleaned[col].mode()[0], inplace=True)

'''
```

# EDA

After cleaning the data, we were ready to conduct further EDA.

```python
# sns style for plots
sns.set(style="whitegrid")

# numerical rating columns for univariate analysis
rating_cols = ['Overall Rating', 'Seat Comfort', 'Cabin Staff Service',
               'Food & Beverages', 'Ground Service', 'Value For Money',
               'Inflight Entertainment', 'Wifi & Connectivity']

# distributions of numerical rating columns
plt.figure(figsize=(15, 10))
for i, col in enumerate(rating_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data_cleaned[col], kde=True, bins=10)
    plt.title(col)
    plt.xlabel('Rating')
    plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# correlation heatmap of numerical ratings
correlation_matrix = data_cleaned[rating_cols].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Ratings")
plt.show()
```

```python
# overall ratings across different Type of Traveller categories
plt.figure(figsize=(12, 6))
sns.boxplot(x='Type Of Traveller', y='Overall Rating', data=data_cleaned)
plt.xticks(rotation=45)
plt.title('Overall Rating by Type Of Traveller')
plt.show()
```

```python
# overall ratings by different Seat Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='Seat Type', y='Overall Rating', data=data_cleaned)
plt.xticks(rotation=45)
plt.title('Overall Rating by Seat Type')
plt.show()
```

```python
from wordcloud import WordCloud

# word cloud for the 'Comment' column
text = " ".join(review for review in data_cleaned.Comment)

# word cloud image
wordcloud = WordCloud(background_color="white", max_words=100).generate(text)

# display wordcloud image
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Word Cloud for Comments")
plt.show()
```

```python
# average ratings for each numerical category
average_ratings = data_cleaned[['Seat Comfort', 'Cabin Staff Service', 'Food & Beverages',
                                'Ground Service', 'Value For Money', 'Inflight Entertainment',
                                'Wifi & Connectivity']].mean()

# bar chart
plt.figure(figsize=(12, 6))
average_ratings.plot(kind='bar')
plt.title('Average Ratings for Each Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# resampling the data to monthly averages
monthly_average = data_cleaned.resample('M', on='Date Published')[['Overall Rating',
                                                                   'Seat Comfort',
                                                                   'Cabin Staff Service',
                                                                   'Food & Beverages',
                                                                   'Ground Service',
                                                                   'Value For Money',
                                                                   'Inflight Entertainment',
                                                                   'Wifi & Connectivity']].mean()

# plotting the average ratings over time
plt.figure(figsize=(15, 8))
monthly_average.plot(ax=plt.gca())
plt.title('Average Ratings Over Time (Monthly)')
plt.ylabel('Average Rating')
plt.xlabel('Date Published')
plt.legend(loc='upper left')
plt.show()
```

```python
# note - the "0's" here are actually mostly NAs, I believe

# rating categories
rating_categories = ['Seat Comfort', 'Cabin Staff Service',
                     'Food & Beverages', 'Ground Service', 'Value For Money',
                     'Inflight Entertainment', 'Wifi & Connectivity']

# rating order
rating_order = [0, 1, 2, 3, 4, 5]

# data for the stacked bar chart
stacked_data = pd.DataFrame(index=rating_order)
for category in rating_categories:
    # Calculating percentage for each rating in the category in the specified order
    category_percentages = data_cleaned[category].value_counts(normalize=True).reindex(rating_order) * 100
    stacked_data[category] = category_percentages

# transpose the data
stacked_data = stacked_data.T.fillna(0)

# stacked horizontal bar chart
stacked_data.plot(kind='barh', stacked=True, figsize=(15, 8))
plt.title('Percentage Distribution of Review Ratings per Category')
plt.xlabel('Percentage of Ratings')
plt.ylabel('Category')
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Summary

The dataset comprised of 2249 entries across 21 columns, including a variety of data types such as integer, float, and object. Unnecessary columns such as 'Unnamed: 0' were dropped. Date fields were converted to datetime objects, facilitating time-based analyses.

We then performed a number of univariate and multivariate analyses of the data to gain useful insights before conducting our model training and testing.


### Rating Distributions:

*   The histograms of ratings showed that 'Overall Rating' and 'Cabin Staff Service' generally received high scores, with many ratings clustered around the upper scale (8-10).
* 'Inflight Entertainment' and 'Wifi & Connectivity' had sparser data but tended to receive lower ratings, indicating less satisfaction in these areas.

### Correlation Analysis:

* The heatmap revealed a strong correlation between 'Cabin Staff Service' and 'Overall Rating' (0.89), suggesting that interactions with cabin staff are a significant determinant of overall passenger satisfaction.

* Moderate correlations were observed between 'Seat Comfort' and 'Overall Rating' (0.65), highlighting comfort as another key factor.

### Comparative Analysis by Traveler Type and Seat:

* Box plots by 'Type of Traveller' showed that business travelers generally rated their experience higher than leisure travelers, with median ratings close to 9 compared to 7 for leisure travelers.

* Analysis by 'Seat Type' indicated that premium seats (Business/First Class) typically received higher ratings compared to Economy, with fewer low scores and a higher median (around 9 for Premium vs. 6 for Economy).

### Word Cloud:

* Frequent words like "comfortable", "efficient", and "friendly" were prominent, suggesting positive feedback in these areas.
* Words like "delay" and "cramped" also appeared, indicating areas of frequent complaints.

### Trends Over Time:

The line graph of monthly averages showed a notable improvement in ratings around mid-2023, possibly linked to service enhancements or seasonal effects.
A slight decline in some categories towards the end of 2023 suggested areas where attention might be needed.

### Conclusion:

Ultimately, this detailed EDA offers comprehensive insights into the Ryanair customer experience dataset through various analytical techniques, from basic statistical summaries to complex visualizations of trends and correlations, which will allow us to accurately build out our model in the next phase of our project.
